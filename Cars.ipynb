```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Assuming 'datasets' is your DataFrame
X = datasets[datasets.columns[:-1]]

# Replace non-numeric values with NaN
X = X.apply(pd.to_numeric, errors='coerce')

# Fill NaN values with a specific numeric value (e.g., 0)
X = X.fillna(0)

# Perform k-means clustering
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
         
```

```python
kmeans =KMeans (n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(X)
X = X.values
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0,1],s=100,c='cyan',label='Toyota')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1,1],s=100,c='yellow',label='Nissan')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2,1],s=100,c='pink',label='Honda')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='magenta',label='Centroids')
plt.title('Clusters of car make')
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], X[y_kmeans == 0, 2], s=100, c='cyan', label='Toyota')
ax.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], X[y_kmeans == 1, 2], s=100, c='yellow', label='Nissan')
ax.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], X[y_kmeans == 2, 2], s=100, c='pink', label='Honda')
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s=100, c='magenta', label='Centroids')

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
plt.title('Clusters of car make (3D)')
plt.legend()
plt.show()
```
